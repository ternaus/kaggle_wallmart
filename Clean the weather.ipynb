Some of the quantities that I have here should be nonnegative. But regression models will return negative values. Thus I will apply log(1+x) to initial features. So that Regression models may work freely

```python
import math

for column in ['heat', 'cool', 'snowfall', 'preciptotal', 'stnpressure', 'sealevel', 'resultspeed', 'resultdir', 'avgspeed', 'day_length']:
    weather[column] = weather[column].apply(lambda x: math.log(1 + x))
```
