# I need to clean weather file.

```python
#matplotlib inline
import pandas as pd
import numpy as np
import os
```

```python
weather = pd.read_csv(os.path.join("data", "weather_modified.csv"))
```

```python
weather.head()
```

```python
weather = weather.drop("Unnamed: 0", 1)
```

I will assume that year is not that important, because of the periodicity, thus I need to extract month and day from date.

```python
def get_month(x):
    return int(x.split("-")[1])
```

```python
weather["month"] = weather["date"].apply(get_month)
```

```python
def get_day(x):
    return int(x.split("-")[2])
```

```python
weather["day"] = weather["date"].apply(get_day)
```

Sunset and sunrize are represented as an int that joins hours and minutes. I would think that length of the light day is important, but actual sunset sunrize time is not.

```python
import datetime
# import time
```

```python
def day_length(x):    
    sunset = x["sunset"]
    sunrise = x["sunrise"]    
    
    if np.isnan(sunset) or np.isnan(sunrise):        
        return np.nan
    else:
        if (str(int(sunset))[-2:]) == '60':
            sunset -= 1
        if (str(int(sunrise))[-2:]) == '60':
            sunrise -= 1               
        b = datetime.datetime.strptime(str(int(sunset)), "%H%M")
        a = datetime.datetime.strptime(str(int(sunrise)), "%H%M")                    
        return (b - a).total_seconds() / 60
```

```python
weather["day_length"] = weather.apply(day_length, axis=1)
```

```python
#drop sunset, sunrise
weather = weather.drop(["sunset", "sunrise"], 1)
```

```python
weather.info(verbose=False)
```

```python
#fill weather with mean values
for column in weather.columns:
    if sum(weather[column].isnull()) > 0:
        mean_column = weather[column].mean()
        weather[column] = weather[column].fillna(mean_column)
```

```python
#save to file

weather.to_csv(os.path.join("data", "weather_mean.csv"), index=False)
```
